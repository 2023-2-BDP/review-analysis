```python
# content, sentiment 컬럼만 추출
train_df = train_df[['content', 'sentiment']]
valid_df = valid_df[['content', 'sentiment']]
test_df = test_df[['content', 'sentiment']]
```

```python
train_df = train_df[train_df['sentiment'].isin([0, 1])]
valid_df = valid_df[valid_df['sentiment'].isin([0, 1])]
test_df = test_df[test_df['sentiment'].isin([0, 1])]
```

```python
# 전처리 함수 정의

from konlpy.tag import Okt
okt = Okt()

stopwords = pd.read_csv("https://raw.githubusercontent.com/yoonkt200/FastCampusDataset/master/korean_stopwords.txt").values.tolist()
# 리뷰 데이터에 많이 포함된 불용어 추가
stopwords.extend(['쿠팡', '리뷰'])

# 텍스트 정제 함수
def text_cleaning(text):

    # 한글과 띄어쓰기를 제외한 모든 부분을 제거
    hangul = re.compile('[^ ㄱ-ㅣ 가-힣]') 
    result = hangul.sub('', text) 

    # okt 객체를 활용해 형태소 단위로 나눈다.
    result = [word for word in okt.nouns(result) if len(word) > 1]
    result = [word for word in result if not word in stopwords]
    return result
```

```python
# 전처리 함수 적용

train_df['content'] = train_df['content'].apply(text_cleaning)
valid_df['content'] = valid_df['content'].apply(text_cleaning)
test_df['content'] = test_df['content'].apply(text_cleaning)
```

```python
train_df.to_csv('data_konlpy_processing_finished/train.csv', index=False, encoding='utf-8', sep='\t')
valid_df.to_csv('data_konlpy_processing_finished/valid.csv', index=False, encoding='utf-8', sep='\t')
test_df.to_csv('data_konlpy_processing_finished/test.csv', index=False, encoding='utf-8', sep='\t')
```
